```python
TwentySecs["LIT101"].plot(subplots=True, figsize=(6,5), legend=True, layout=(-1, 1), sharex=True)
plt.xlabel("Time")
plt.suptitle("Time Series Plots for Each Sensor", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout to make room for title
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Plotting the data
axes = data["LIT101"].plot( figsize=(6,5), legend=True, layout=(-1, 1), sharex=True)

# # If there is only one subplot, axes is a 1D array with a single element, so we access axes[0]
# ax = axes[0] if isinstance(axes, np.ndarray) else axes

# Adding vertical lines
axes.axvline(x=10000, color='red', linestyle='--')  # Vertical line at the 5th data point
axes.axvline(x=20000, color='red', linestyle='--')  # Vertical line at the 10th data point

# Optionally, you can shade the area between the 5th and 10th data point using axvspan
axes.axvspan(10000, 20000, color='yellow', alpha=0.3)  # Highlight the range from 5 to 10

# Show the plot
plt.show()
```
